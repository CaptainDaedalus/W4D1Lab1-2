```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Import the necessary libraries if you are starting a new notebook. Using the same data as the previous lab: 
#we_fn_use_c_marketing_customer_value_analysis.csv
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
display(data.head())
display(data.shape)
display(data.info)
```

```python
#2. Find all of the categorical data. Save it in a categorical_df variable.
categorical_df = data.select_dtypes(include=['object'])
display(categorical_df)
```

```python
#3. Check for NaN values.
categorical_df.isna().sum()
```

```python
#4. Check all unique values of columns.
for col in categorical_df:
    print(categorical_df[col].unique())
```

```python
categorical_df['Customer'].shape
#I suppose it makes sense for there to be so many (unique?) customer id values.
```

```python
#5. Check dtypes. Do they all make sense as categorical data? Yes, I suppose so, with the exception of Effective To Date maybe.
#categorical_df.dtypes
categorical_df.dtypes
```

```python
#6. Does any column contain alpha and numeric data? Decide how to clean it.
# Yes "Customer", "Policy" and "Renew Offer Type" columns contain alpha and numerica data.
categorical_df['Renew Offer Type']=categorical_df['Renew Offer Type'].str.replace('1', '-One', regex=True).str.replace('2', '-Two', regex=True).str.replace('3', '-Three', regex=True).str.replace('4', '-Four', regex=True)
categorical_df['Policy']=categorical_df['Policy'].str.replace('1', '-One', regex=True).str.replace('2', '-Two', regex=True).str.replace('3', '-Three', regex=True)
display(categorical_df)
#replaced the digit with the word for the digit
```

```python
#7. Would you choose to do anything else to clean or wrangle the categorical data? Comment your decisions. 

#I want to reduce the number of unique 'Effective to Date' values by grouping them
```

```python
#8. Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?
compare = categorical_df[['Policy Type', 'Policy']]
display(compare)

#What is important is that 'Corporate' and/or 'Personal' always match each other in the 'Policy Type' and 'Policy' columns. 
#Furthermore, for each type of the 3 policies, there are 3 sub-policies, L-One, L-Two, and L-Three.
```
