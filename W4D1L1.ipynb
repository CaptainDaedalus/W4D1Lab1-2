```python
numerical = data.select_dtypes(include=np.number)
```

```python
numerical
```

```python
numerical['Income'].unique()
```

```python
def separate_num(df, thr=10):
    cont = []
    dis = []
    for col in df.columns:
        if df[col].nunique() < thr:
            dis.append(col)
        else:
            cont.append(col)
            
    cont_df = df[cont]
    dis_df  = df[dis]
    
    return cont_df, dis_df
```

```python
continuous, discrete = separate_num(numerical)
```
